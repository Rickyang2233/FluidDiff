```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
with h5py.File("../data/2d_full_smoke/NSFullSmoke2D_resolution_64x64_train.h5", 'r') as f:
    u = f['2DNSFullSmoke']['u'][:]
    vx = f['2DNSFullSmoke']['vx'][:]
    vy = f['2DNSFullSmoke']['vy'][:]
    t = f['2DNSFullSmoke']['t'][:]
    u0 = f['2DNSFullSmoke']['u_init'][:]
```

```python
u0.shape
```

```python
fig, ax = plt.subplots(5, 5, figsize=(15, 15))
for i in range(5):
    for j in range(5):
        ax[i, j].imshow(u[i*5, j*10, :, :])
        ax[i, j].set_title(f"t={t[i*5, j*10]:.2f}")
        ax[i, j].set_xticks([])
        ax[i, j].set_yticks([])
plt.show()
```

```python
fig, ax = plt.subplots(1, 5, figsize=(15, 3))
for i in range(5):
    ax[i].imshow(u0[i*5, :, :])
    ax[i].set_axis_off()
```
