```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = h5py.File('ns_floating_sim_2d-seed1.h5', 'a')
```

```python
for k in f.keys():
    print(k, f[k].shape)
```

```python
density = f['density'][:]
velocity = f['velocity'][:]
pressure = f['pressure'][:]
initial_density = f['initial_density'][:]
t = f['t'][:]
```

```python
n = np.random.choice(1000)
fig, ax = plt.subplots(4, 5, figsize=(18, 12))
for i in range(1, 6):
    ax[0, i-1].imshow(density[n, i*8-1, :, :, 0], origin='lower')
    ax[1, i-1].imshow(velocity[n, i*8-1, :, :, 0], origin='lower')
    ax[2, i-1].imshow(velocity[n, i*8-1, :, :, 1], origin='lower')
    ax[3, i-1].imshow(pressure[n, i*8-1, :, :, 0], origin='lower')
plt.show()
```
