```python
from phi.torch.flow import *
```

```python
noise = Noise((batch(n_exp=3)), scale=4.0, smoothness=1.5)
```

```python
u = abs(
    CenteredGrid(
        noise,
        extrapolation.BOUNDARY,
        x=64,
        y=64,
        bounds=Box(x=16.0, y=16.0)
    )
)
```

```python
plot(u, show_color_bar=False)
```
