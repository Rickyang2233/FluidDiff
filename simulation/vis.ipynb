```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_range = np.arange(2, 14+0.125, 0.125)
y_range = np.arange(2, 14+0.125, 0.125)
xx, yy = np.meshgrid(x_range, y_range)
points = [[x, y] for x, y in zip(xx.flatten(), yy.flatten())]
```

```python
np.random.seed(215)
```

```python
dt = 0.05
total_time = 15.0
start_time = 5.0
n_total_steps = int(total_time / dt)
n_start_steps = int(start_time / dt)
n_frames_stored = 10
n_steps_per_frame = (n_total_steps - n_start_steps) // n_frames_stored
```

```python
n_total_steps, n_start_steps, n_frames_stored, n_steps_per_frame
```

```python
time_series = np.arange(1, n_total_steps+1) * dt
```

```python
for i in range(1, n_total_steps+1):
    if i >= n_start_steps and (i - n_start_steps) % n_steps_per_frame == 0:
        index = (i - n_start_steps) // n_steps_per_frame
        print(f"t={time_series[i-1]}, index={index}")
```
