```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def func(ep):
    return ((1 + math.cos(ep * math.pi / 300)) / 2) * (1 - 0.1) + 0.1
```

```python
eps = np.arange(0, 300, 1)
plt.plot(eps, [func(ep) for ep in eps])
```
